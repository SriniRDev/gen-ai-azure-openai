### Normalise the vector
Above embedding gives a vector with many dimensions which can be reduced using the below function

```python
import numpy as np
def normalise_l2(x):
    x = np.array(x)
    if x.ndim == 1:
        norm = np.linalg.norm(x)
        if norm == 0:
            return x
        return x/norm
    else:
        norms = np.linalg.norm(x,2, axis=1, keepdims = True)
        return np.where(norm == 0,x,x/norm)
```

```python
cut_dim = response.data[0].embedding[:7]
norm_dim = normalise_l2(cut_dim)
print(norm_dim)
```
